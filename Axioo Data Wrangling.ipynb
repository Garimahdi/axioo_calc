# Calculating Cost-per-page for CMYK

```python
util = pd.read_excel(
    r"D:\\Astragraphia\\BO1\\Axioo\\Assessment Data\\axioo_assessment_rev.xlsx",
    sheet_name = 'Util Yield'
)
cost = pd.read_excel(
    r"D:\\Astragraphia\\BO1\\Axioo\\Assessment Data\\axioo_assessment_rev.xlsx",
    sheet_name = 'Toner-Other Costs'
)

u = list(util.columns)
c = list(cost.columns)
```

```python
util.info()
```

```python
cost.info()
```

```python
costCol = [c[3],c[4],c[5],c[6],c[7]]
util = util.join(cost[costCol], how = 'left')
```

```python
util.head()
```

```python
util['CPP BW_Theo'] = util[c[6]]/util[u[7]]
util['CPP CLR_Theo'] = 0

for x in util[c[7]]:
    if x == 0:
        util['CPP CLR_Theo'] = (util[c[3]] + util[c[4]] + util[c[5]])/util[u[4]]
    else:
        util['CPP CLR_Theo'] = util[c[3]]/util[u[4]]
```

```python
util['CPP CLR_Theo'].fillna(0, inplace=True)
util['CPP Mix_Theo'] = util['CPP BW_Theo'] + util['CPP CLR_Theo']
util.head()
```

```python
util['Total Cost'] = (util['CPP Mix_Theo'] * util[u[8]] * 4).astype(int)
util.drop(columns = [c[3],c[4],c[5],c[6],c[7],'CPP Mix_Theo'], inplace = True)
util['CPP CLR_Theo'].fillna(0, inplace = True)
```

```python
util.to_excel(
    "D:\\Astragraphia\\BO1\\Axioo\\Assessment Data\\axioo_assessment_rev_v2.xlsx",
    sheet_name = 'Utilisation',
    index = False
)
```

```python
path = r"D:\\Astragraphia\\BO1\\Axioo\\Assessment Data\\axioo_assessment_rev_v2.xlsx"
with pd.ExcelWriter(path, engine='openpyxl') as writer:
    util.to_excel(writer, sheet_name = 'Utilisation', index = False)
    cost.to_excel(writer, sheet_name = 'TOC', index = False)
```
