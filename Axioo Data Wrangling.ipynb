```python
import pandas as pd
import numpy as np
import openpyxl
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# Identify Sheet Names in the File

filePath = r"D:\Astragraphia\BO1\Axioo\Assessment Data\Copy of Control Asset Printer Maret Mei 2023.xlsx"
daFile = openpyxl.load_workbook(filePath)
daSheet = daFile.sheetnames
daFile.sheetnames
```

```python
daList = pd.read_excel(
    filePath,
    sheet_name = daSheet[0]
)
daList.head()
```

```python
#daList = daList.iloc[1:len(daList)-1]
daList.head()
try:
    print(list(daList[daList.columns[1]].iloc[0:1])[0])
except Exception as e:
    print(e)

dOld = list(daList.columns)
dNew = []
for a in range(0,len(dOld)):
    dNew.append(list(daList[dOld[a]].iloc[0:1])[0])
```

```python
daList.head()
```

```python
print(dNew)
print(dOld)

for o, n in zip(dOld, dNew):
    daList.rename(columns={o:n}, inplace = True)
daList.drop(index = 1, inplace = True)
```

```python
#daList = daList.reset_index()
#daList.drop(columns = ['index'], inplace = True)
d = list(daList.columns)
daList[[d[1],d[2]]].fillna(method = 'ffill', inplace = True)
daList.to_excel("D:\\Astragraphia\\BO1\\Axioo\\Assessment Data\\Asset Printer Maret Mei 2023.xlsx")
daList.to_csv("D:\\Astragraphia\\BO1\\Axioo\\Assessment Data\\Asset Printer Maret Mei 2023.csv")
```
